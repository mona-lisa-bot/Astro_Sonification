```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#Load dataset from the .npy file
data = np.load('images.npy')
```

```python
data.shape
```

```python
#data is 4D array of (n_images, height, width, channels)
image = data[7]
image.shape
```

```python
#Checking if the image is RGB format or BGR format
```

```python
#Display the image using OpenCV
cv2.imshow('Galaxy Image', image) #This will show the BGR image
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display the image without conversion
plt.imshow(image)
plt.title('Original_image')
plt.axis('off')
plt.show()

# Convert to RGB format (if needed)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with proper RGB conversion
plt.imshow(rgb_image)
plt.title('With RGB Conversion')
plt.axis('off')
plt.show()
```

```python
print(image[0, 0])  # Prints the pixel value at the top-left corner
```

```python
#image is in RGB format only
```

```python
# Extract pixel data (RGB values)
height, width, channels = image.shape
pixel_data = image.reshape(height * width, channels)
```

```python
# Store pixel data in a CSV file
df = pd.DataFrame(pixel_data, columns=['R', 'G', 'B'])
df.to_csv('galaxy_image_pixel_data.csv', index=False)
print('Pixel data saved to galaxy_image_pixel_data.csv')
```

```python
#BGR conversion
bgr_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
# Visualization using matplotlib: display the image
plt.figure(figsize=(6, 6))
plt.imshow(bgr_image)
plt.title('Galaxy Image(BGR)')
plt.axis('off')
plt.show()
```

```python
# Visualization using a heatmap of intensity (for grayscale visualization)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6, 6))
plt.imshow(gray_image, cmap='hot', interpolation='nearest')
plt.title('Galaxy Image Intensity Heatmap')
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
# Visualize individual RGB channels (e.g., Red channel)
plt.figure(figsize=(6, 6))
plt.imshow(image[:, :, 0], cmap='Reds')  # Visualize the Red channel
plt.title("Red Channel Intensity")
plt.axis('off')
plt.show()
```

```python
#add histograms and scatter plots
```

```python
# Image is in RGB format
red, green, blue = image[:, :, 0], image[:, :, 1], image[:, :, 2]

# Plot histograms for each channel
plt.figure(figsize=(10, 5))

# Red Channel
plt.subplot(1, 3, 1)
plt.hist(red.ravel(), bins=256, color='red', alpha=0.7)
plt.title('Red Channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.ylim(0, 3000)  

# Green Channel
plt.subplot(1, 3, 2)
plt.hist(green.ravel(), bins=256, color='green', alpha=0.7)
plt.title('Green Channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.ylim(0, 3000) 

# Blue Channel
plt.subplot(1, 3, 3)
plt.hist(blue.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Blue Channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.ylim(0, 3000) 

plt.tight_layout()
plt.show()

```

```python
# Reshape the image to a 2D array of pixels and their colors
pixels = image.reshape(-1, 3)  # Shape (n_pixels, 3)

# Plot relationships between channels
plt.figure(figsize=(10, 5))

# Red vs Green
plt.subplot(1, 3, 1)
plt.scatter(pixels[:, 0], pixels[:, 1], color='yellow', s=1, alpha=0.5)
plt.xlabel('Red')
plt.ylabel('Green')
plt.title('Red vs Green')

# Red vs Blue
plt.subplot(1, 3, 2)
plt.scatter(pixels[:, 0], pixels[:, 2], color='magenta', s=1, alpha=0.5)
plt.xlabel('Red')
plt.ylabel('Blue')
plt.title('Red vs Blue')

# Green vs Blue
plt.subplot(1, 3, 3)
plt.scatter(pixels[:, 1], pixels[:, 2], color='cyan', s=1, alpha=0.5)
plt.xlabel('Green')
plt.ylabel('Blue')
plt.title('Green vs Blue')

plt.tight_layout()
plt.show()
```
